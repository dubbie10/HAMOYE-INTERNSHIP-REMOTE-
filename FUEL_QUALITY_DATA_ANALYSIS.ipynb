## CLEANING THE DATA
___

### Cleaning the 'fuel_unit' column

```python
fuel_data.groupby('fuel_type_code_pudl')['fuel_unit'].value_counts()
```

We can see that each fuel type is measured in different units.

I could choose to replace the missing 'fuel_unit' values based on the most common units of the corresponding 'fuel_type_code_pudl' but the probability, although small, that the 'fuel_unit' might be mismatched with the respective quantites cannot be overlooked.

So I will be opting to remove the rows with null 'fuel_unit' values in the code block below to make my analysis more accurate.

```python
fuel_data.dropna(inplace=True)
```

```python
fuel_data.isnull().sum()
```

Null values no longer exist in our dataframe.

### Cleaning the 'fuel_mmbtu_per_unit' column

* I'll check for outliers using boxplot and then remove them.

```python
fig, ax=plt.subplots(figsize=(10,7))
ax.boxplot(x=fuel_data['fuel_mmbtu_per_unit'])
ax.set_xlabel('fuel_mmbtu_per_unit')
ax.set_title('FUEL_MMBTU_PER_UNIT_DISTRIBUTION')
ax.set_ylim(-5,50)

plt.show()
```

### Observation

* Majority of the 'fuel_mmbtu_per_unit' is distributed between around 0 and 40.

### Action

* I will retain rows that have 'fuel_mmbtu_per_unit' values of between 0 and 40.

```python
fuel_data = fuel_data[fuel_data['fuel_mmbtu_per_unit'].between(0,40)] #keeps only rows with values of "fuel_mmbtu_per_unit" column at 0-40.
```

### Cleaning the 'fuel_cost_per_mmbtu' column

* I'll check for outliers using boxplot and then remove them.

```python
fig, ax=plt.subplots(figsize=(10,7))
ax.boxplot(x=fuel_data['fuel_cost_per_mmbtu'])
ax.set_xlabel('fuel_cost_per_mmbtu')
ax.set_title('FUEL_COST_PER_MMBTU_DISTRIBUTION')
ax.set_ylim(-20,20)
plt.show()
```

### Observation

* Majority of the 'fuel_cost_per_mmbtu' is distributed between around -10 and 16.

### Action

* I will retain rows that have 'fuel_cost_per_mmbtu' values of between 0 and 16, since it is unrealistic to have a cost below 0.

```python
fuel_data = fuel_data[fuel_data['fuel_cost_per_mmbtu'].between(0,16)] #keeps only rows with values of "fuel_cost_per_mmbtu" column at 0-16.
```

### Which fuel has the most energy content?

BTU standard unit of measurement used to denote both the amount of heat energy in fuels. A BTU is the amount of heat required to increase the temperature of a pint of water (which weighs exactly16 ounces) by one degree Fahrenheit. 

mmbtu which refers to a thousand thousand (million) BTU is occasionally used as a standard unit of measurement for natural gas and provides a convenient basis for comparing the energy content of various grades of natural gas and other fuels.

In this section, I will look to determine the strongest type of fuel.

## FUEL CONSUMPTION TRENDS

To be able to accurately calculate this data, I am going to create a new column for fuel quantity burned based on a single unit.

This is because the fuel_qty_burned column is based on nine(9) different units, therefore it'll be more accurate to compare them based on one single unit.

The fuel mmbtu per unit signifies the mmbtu rating for each unit of the diffrent fuel units. Multiplying this by the corresponding 'fuel_quantity_burned' will give the fuel_quantity_burned converted to mmbtu, therebey harmonizing everything to one similar unit for basis of comparison. 

```python
fuel_qty_burned_mmbtu = fuel_data['fuel_qty_burned']*fuel_data['fuel_mmbtu_per_unit'] #creates a new series.
fuel_data['fuel_qty_burned_mmbtu'] = fuel_qty_burned_mmbtu #adds a new column 'fuel_qty_burned_mmbtu' to the data frame.
```

```python
fuel_data.head(3)
```

We can see our new 'fuel_qty_burned_mmbtu' column in the dataframe.

### Is the fuel price dependent on it's consumption?
